```python
url = "https://elpais.com/us/migracion/2025-04-18/el-gobierno-contraataca-en-el-caso-de-kilmar-abrego-garcia-de-deportarlo-por-error-a-pintarlo-como-un-monstruo.html"

import requests

res = requests.get(url)
res.raise_for_status()
print(res.text[:1000])  # Print the first 1000 characters of the response
```

```python
from markitdown import MarkItDown

md = MarkItDown(enable_plugins=True)
result = md.convert("../dummy/noticia.html")
print(result)
```

```python
from markitdown import MarkItDown

md = MarkItDown(enable_plugins=True)
result = md.convert(res)

# remove duplicated HTML tags
import re
result_md = re.sub(r"<(h\d+)>", r"<\1>", result.markdown)

print(result.title)
print(result_md)
```
