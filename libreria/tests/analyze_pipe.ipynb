```python
verified_count = len([w for w in pipe.search_webpages if w.veredict == "verified"])
unverified_count = len([w for w in pipe.search_webpages if w.veredict == "unverified"])
unrelated_count = len([w for w in pipe.search_webpages if w.veredict == "unrelated"])
total_count = len(pipe.search_webpages)

print("verified:", verified_count)
print("unverified:", unverified_count)
print("unrelated:", unrelated_count)
print("Total:", total_count)

if total_count > 0:
    print(f"verified (%): {verified_count / total_count * 100:.2f}%")
    print(f"unverified (%): {unverified_count / total_count * 100:.2f}%")
    print(f"unrelated (%): {unrelated_count / total_count * 100:.2f}%")
else:
    print("No webpages to calculate percentages.")
```
