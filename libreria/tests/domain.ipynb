Geolocation

```python
import requests

def get_geolocation(domain_or_ip):
    response = requests.get(f'https://ipinfo.io/{domain_or_ip}/json')
    if response.status_code == 200:
        data = response.json()
        #print(data)
        return {
            'ip': data.get('ip'),
            'city': data.get('city'),
            'region': data.get('region'),
            'country': data.get('country'),
            'loc': data.get('loc')  # latitude,longitude
        }
    else:
        return {'error': f"Failed to retrieve data: {response.status_code}"}

# Example usage:
info = get_geolocation(ip)  # or 'google.com'
print(info)
```
