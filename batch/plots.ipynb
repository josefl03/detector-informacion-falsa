Fake News Corpus Spanish

```python
import matplotlib.pyplot as plt

FNT_WEIGHT = 'normal'
FNT_SIZE = 16
FNT_SIZE2 = 12

# Data
labels = ['Aciertos', 'Fallos']
counts = [93, 10]
percentages = [90.29, 9.71]
# Soft (pastel) colors
#colors = ["#0066ff", "#ffbe0b"]  # Bright blue and vibrant yellow-orange for more color contrast
colors = ["#1de74f", "#f04c4c"]  # Green for OK, pink for Blocked

# Plot with less horizontal aspect ratio and thinner bars
fig, ax = plt.subplots(figsize=(5, 6))  # More vertical, less horizontal
bars = ax.bar(labels, counts, color=colors, width=0.7)  # Thinner bars

# Add percentage labels on top of bars with larger font
for bar, pct in zip(bars, percentages):
    height = bar.get_height()
    ax.annotate(f'{pct:.2f}%',
                xy=(bar.get_x() + bar.get_width() / 2, height),
                xytext=(0, 3),  # 3 points vertical offset
                textcoords="offset points",
                ha='center', va='bottom',
                fontsize=FNT_SIZE2, fontweight=FNT_WEIGHT)

#ax.set_ylabel('Count', fontsize=FNT_SIZE, fontweight=FNT_WEIGHT)
#ax.set_xlabel('Category', fontsize=FNT_SIZE, fontweight=FNT_WEIGHT)
ax.set_xticklabels(labels, fontsize=FNT_SIZE, fontweight=FNT_WEIGHT)
ax.set_title('Resultados del An√°lisis', fontsize=FNT_SIZE+2, fontweight=FNT_WEIGHT)
ax.tick_params(axis='x', labelsize=FNT_SIZE)
ax.tick_params(axis='y', labelsize=FNT_SIZE)
ax.grid(True, axis='y', linestyle='--', alpha=0.6)  # Enable grid lines on y-axis
plt.show()
```

```python
import matplotlib.pyplot as plt

FNT_WEIGHT = 'normal'
FNT_SIZE = 16
FNT_SIZE2 = 12

# Data
labels = ['Sociedad', 'Ciencia', 'Deporte', 'Salud', 'Educaci√≥n', 'Seguridad', 'Econom√≠a']
counts = [46, 13, 8, 7, 10, 5, 4]
percentages = [92.00, 100.00, 80.00, 100.00, 83.33, 100.00, 66.67]

# Ordenar de mayor a menor seg√∫n percentages
sorted_indices = sorted(range(len(percentages)), key=lambda i: percentages[i], reverse=True)
labels = [labels[i] for i in sorted_indices]
counts = [counts[i] for i in sorted_indices]
percentages = [percentages[i] for i in sorted_indices]
colors = [colors[i] for i in sorted_indices]

# Distinct, visually appealing colors for each category
import matplotlib.colors as mcolors

def saturate_color(hex_color, saturation_factor=99):
    rgb = mcolors.hex2color(hex_color)
    hsv = mcolors.rgb_to_hsv(rgb)
    hsv = list(hsv)
    hsv[1] = min(hsv[1] * saturation_factor, 1.0)
    saturated_rgb = mcolors.hsv_to_rgb(hsv)
    return mcolors.to_hex(saturated_rgb)

# Manually shuffled colors for better visual distinction between adjacent bars
colors = ["#3a86ff", "#e71b1b", "#59b800", "#df53d3", "#fb5607", "#00bcd4", "#dee200"]

fig, ax = plt.subplots(figsize=(12, 6))
bars = ax.bar(labels, percentages, color=colors, width=0.7)

ax.set_xticklabels(labels, fontsize=FNT_SIZE, fontweight=FNT_WEIGHT)
ax.set_title('Resultados por categor√≠a en el "Fake News Corpus Spanish"', fontsize=FNT_SIZE+2, fontweight=FNT_WEIGHT)
ax.tick_params(axis='x', labelsize=FNT_SIZE)
ax.tick_params(axis='y', labelsize=FNT_SIZE)
ax.grid(True, axis='y', linestyle='--', alpha=0.6)
plt.show()
```
